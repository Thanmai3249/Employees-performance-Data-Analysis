```python
#9 . average weekly response time for each agent

performance[['response_hours','response_minutes','response_seconds']]=performance["Average Response Time"].str.split(":",expand=True)

performance['response_hours']=performance['response_hours'].astype(int)
performance['response_minutes']=performance['response_minutes'].astype(int)
performance['response_seconds']=performance['response_seconds'].astype(int)

performance["Response time in seconds"]=(performance['response_hours']*3600) + (performance['response_minutes']*60) + (performance['response_seconds'])
performance.head()
```

```python
Average_Response_Time = performance.groupby(['Agent Name'])['Response time in seconds'].sum()/5
Average_Response_Time.sort_values(ascending = False)
```

```python
#9 . average weekly response time for each agent

performance[['resolution_hours','resolution_minutes','resolution_seconds']]=performance["Average Resolution Time"].str.split(":",expand=True)

performance['resolution_hours']=performance['resolution_hours'].astype(int)
performance['resolution_minutes']=performance['resolution_minutes'].astype(int)
performance['resolution_seconds']=performance['resolution_seconds'].astype(int)

performance["Resolution time in seconds"]=(performance['resolution_hours']*3600) + (performance['resolution_minutes']*60) + (performance['resolution_seconds'])

Average_Resolution_Time = performance.groupby(['Agent Name'])['Resolution time in seconds'].sum()/5
Average_Resolution_Time.sort_values(ascending = False)
```
